```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def linearRegression(x,y,figName):
    w,b=0,0
    m=len(x)
    alpha=0.01
    iterations=1000
    error=np.zeros(iterations)
    for i in range(iterations):
        y_wb = w*x+b
        dw=(1/m)*np.sum(x*(y_wb-y))
        db=(1/m)*np.sum(y_wb-y)
        w,b=w-alpha*dw,b-alpha*db
        error[i]=np.sum(np.sqrt(dw**2+db**2))
        print(f'{i+1}) w = {w}\tb = {b}\terror = {error[i]}')
    fig, (errplot, varplot) = plt.subplots(1, 2, figsize=(12, 6))
    errplot.plot(error)
    errplot.set_title('Error')
    errplot.set_xlabel('Iterations')
    errplot.set_ylabel('Error')
    varplot.scatter(x,y,marker='x',c='r')
    varplot.plot(x,w*x+b,label=f'y={w:.3}x+{b:.3}')
    varplot.set_title('Linear Regression')
    varplot.set_xlabel('x')
    varplot.set_ylabel('y')
    fig.suptitle('Error and Linear Regression plot')
    plt.legend()
    plt.savefig(figName, dpi=300)
    plt.show()
```

```python
x  = np.array([-2,1,3])
y = np.array([-1,1,2])
linearRegression(x,y,'q1.png')
```

```python
x  = np.array([-1,0,1,2])
y = np.array([0,2,4,5])
linearRegression(x,y,'q2.png')
```

```python
x = np.array([0,1,2,3,4])
y = np.array([2,3,5,4,6])
linearRegression(x,y,'q3.png')
```
