```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('Practice/data/data.csv')
x = data.iloc[:, 0].values
y = data.iloc[:, 1].values
x.sort();y.sort()
plt.scatter(x,y,marker='x',c='r')
```

```python
alpha = 0.01
iterations = 1000
error = 1e-3
```

```python
def GradientDescentByIterations(x, y, alpha, iterations):
    w, b = 0, 0
    m = len(x)
    for i in range(iterations):
        y_pred = w*x+b
        dw = (1/m)*np.sum(x*(y_pred-y))
        db = (1/m)*np.sum(y_pred-y)
        w = w-alpha*dw
        b = b-alpha*db
        print(f'Iteration No. : {i+1}\n\tw : {w},\tb : {b}\t')
    plt.scatter(x, y, marker='x', c='r')
    plt.plot(x, w*x+b, label='$y = %.2fx + %.2f$' % (w, b))
    plt.legend()
    plt.savefig('oneVariableIterations.png',dpi=300)
```

```python
def GradientdescentByError(x, y, alpha, error):
    w, b = 0, 0
    m = len(x)
    for i in range(1000000000):
        y_pred = w*x+b
        dw = (1/m)*np.sum(x*(y_pred-y))
        db = (1/m)*np.sum(y_pred-y)
        w = w-alpha*dw
        b = b-alpha*db
        print(f'Iteration No. : {i+1}\n\tw : {w},\tb : {b}\terror : {np.sqrt(dw**2+db**2)}')
        if np.sqrt(dw**2+db**2) < error:
            break
    plt.scatter(x, y, marker='x', c='r')
    plt.plot(x, w*x+b, label='$y = %.2fx + %.2f$' % (w, b))
    plt.legend()
    plt.savefig('oneVariableError.png',dpi=300)
```

```python
GradientDescentByIterations(x,y,alpha,iterations)
```

```python
GradientdescentByError(x,y,alpha,error)
```
